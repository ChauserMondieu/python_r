# Second Kind Surface Integral

$$ 2z=x^2+y^2 $$

1. **projection = '3d'** is the key element for matplot plotting 3-dimentional pictures
2. **np.meshgrid()** is the key function for surface generation

```python
from matplotlib import pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(1,1,
                       facecolor="whitesmoke",edgecolor="grey",
                       subplot_kw={"projection":"3d"})
ax.view_init(elev=55,azim=25)

# preparing for each variables
X = np.arange(-5,5,0.25)
Y = np.arange(-5,5,0.25)
X,Y = np.meshgrid(X,Y)
Z = (X**2 + Y**2)/2
P = np.full_like(X,2)

surf = ax.plot_surface(X,Y,Z, cmap=cm.coolwarm,
                       linewidth=0,antialiased=True,
                       label=r'$z=\frac{1}{2}(x^2+y^2)$')

plane = ax.plot_surface(X,Y,P,cmap=cm.autumn,
                        linewidth=0,antialiased=True,
                        label=r'z=2')
line_lim = Z == P
line_X = []
line_Y = []
line_Z = []

for i,b in zip(X.flat,line_lim.flat):
    if(b):
        line_X.append(i)

for i,b in zip(Y.flat,line_lim.flat):
    if(b):
        line_Y.append(i)

for i,b in zip(P.flat,line_lim.flat):
    if(b):
        line_Z.append(i)
        
line = ax.scatter(np.asarray(line_X),
                  np.asarray(line_Y),
                  np.asarray(line_Z),
                  color="cyan",
                  linewidth=100,antialiased=True,
                  label=r'intercept')


# wrong_attribute error for matplotlib processing plot3d with labels
surf._facecolors2d=surf._facecolor3d
surf._edgecolors2d=surf._edgecolor3d

plane._facecolors2d=plane._facecolor3d
plane._edgecolors2d=plane._edgecolor3d

line._facecolors2d=plane._facecolor3d
line._edgecolors2d=plane._edgecolor3d

ax.legend()

# customize z-axis
ax.set_zlim(-10, 10)
ax.zaxis.set_major_locator(LinearLocator(8))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')
# add corlobars
fig.colorbar(surf, shrink=0.4, aspect=5)
fig.show()
```
